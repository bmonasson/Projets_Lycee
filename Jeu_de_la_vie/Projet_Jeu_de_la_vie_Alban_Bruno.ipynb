```python
#Projet Jeu de la vie - Alban Delespaul et Bruno Monasosn



import pygame
import time
import random





#ouverture de la fenêtre pygame
surf = pygame.display.set_mode((500,565))
surf.fill((255,255,255))


#variable globale qui correspond au nombre de cellules du coté de la grille
#(la grille a donc TAILLE**2 cellules au total)
TAILLE = input("Taille de la grille ? ")
#boucle pour qui ne s'arrète pas tant que l'input n'est pas un entier strictement positif
essai = True
while essai :
    try :
        TAILLE = int(TAILLE)
    except :
        TAILLE = input("Merci d'inidiquer un entier strictement positif ")
    else :
        if 0 < TAILLE  :
            essai = False
        else : 
            TAILLE = input("Merci d'inidiquer un entier strictement positif ")


            
            
            
            

class Cellule:
    def __init__(self, position):
        self.position = position
        self.en_vie = False #les cellules sont mortes dans leur configuration initiale

    def statut(self):
        """
        fonction qui tue la cellule ou la fait naitre selon les règles du jeu
        cellule vivante si self.en_vie == True et morte si self.en_vie == False
        """
        if self.en_vie == True : #une cellule en vie meurt si elle n'a pas 2 ou 3 voisins vivants
            if self.nbvoisins != 2 and self.nbvoisins != 3 :
                self.en_vie = False
                coloriage(surf, self.position, self.en_vie) #on change la couleur de la cellule
                return
        if self.en_vie == False: #une cellule morte nait si elle a 3 voisins vivants
            if self.nbvoisins == 3 :
                self.en_vie = True
                if self.position[1] < TAILLE +2 :
                    coloriage(surf, self.position, self.en_vie) #on change la couleur de la cellule
                return

    def voisins(self, grille):
        """
        fonction qui compte le nombre de voisins vivants d'une celulle
        """
        self.nbvoisins = 0
        #celulles voisines du haut :
        for i in range(self.position[1] - 1, self.position[1] + 2):
            if grille[self.position[0] - 1][i].en_vie == True :
                self.nbvoisins += 1

        #cellules voisines du bas :
        for i in range(self.position[1] - 1, self.position[1] + 2):
            if grille[self.position[0] + 1][i].en_vie == True :
                self.nbvoisins += 1

        #cellule voisine de gauche
        if grille[self.position[0]][self.position[1] - 1].en_vie == True :
            self.nbvoisins += 1

        #cellule voisine de droite
        if grille[self.position[0]][self.position[1] + 1].en_vie == True :
            self.nbvoisins += 1
        return


    
    
    
    
    

def affichage_grille(surf):
    """
    affiche une grille de TAILLE x TAILLE cases
    """
    longueur = 500 / TAILLE #longuer = taille de chaque case

    for i in range(TAILLE + 2):
        pygame.draw.line(surf,(0,0,0),(longueur*i, 0),(longueur*i,500),2) #lignes noires
    for j in range(TAILLE + 1):
        pygame.draw.line(surf,(0,0,0),(0, longueur*j),(500, longueur*j),2) #lignes noires

    return




def creation_grille():
    """
    crée la grille de jeu (grille)
    nb : les cellules sont mortes à leur création
    """
    grille = []
    #grille de (TAILLE + 2)x(TAILLE + 2) cases (TAILLE x TAILLE plus 1 colonne/ligne de marge de chaque côté)
    for i in range(TAILLE + 2):
        ligne = []
        for j in range(TAILLE + 2):
            c = Cellule((i, j))
            ligne.append(c) #ajoute cellule morte de coordonées (i,j) à la grille
        grille.append(ligne)
    return grille




def etape(grille):
    """
    simule une étape du jeu de la vie
    """
    #tout d'abord on compte le nb de voisins de chaque celulle :
    for ligne in range(1, TAILLE + 1): #on parcourt uniquement la grille en elle-même et pas les marges
        for colonne in range(1, TAILLE + 1):
            grille[ligne][colonne].voisins(grille)

    #ensuite on met à jour le statut de chaque cellule en fonction du nb de voisins qu'on vient de calculer :
    for ligne in range(1, TAILLE + 1):
        for colonne in range(1, TAILLE + 1):
            grille[ligne][colonne].statut()

    #finalement on met à jour la marge
    mise_a_jour_marges(grille)

    #et on réaffiche la grille car sinon elle disparait sous les cases
    affichage_grille(surf)
    return surf



def mise_a_jour_marges(grille):
    """
    fonction qui met à jour la marge qui se situe tout au tour de la grille
    pour mettre à jour la marge on copie la valeur des cellules auxquelles cette marge correspond
    """
    #mise à jour des 4 coins
    grille[0][0].en_vie = grille[TAILLE][TAILLE].en_vie #coin d'en haut à gauche
    grille[0][TAILLE + 1].en_vie = grille[TAILLE][1].en_vie #coin d'en haut à droite
    grille[TAILLE + 1][0].en_vie = grille[1][TAILLE].en_vie #coin d'en bas à gauche
    grille[TAILLE + 1][TAILLE + 1].en_vie = grille[1][1].en_vie #coin d'en bas à droite

    #mise à jour des 2 lignes de marge
    for i in range(1, TAILLE + 1):
        grille[0][i].en_vie = grille[TAILLE][i].en_vie #marge du haut
        grille[TAILLE + 1][i].en_vie = grille[1][i].en_vie #marge du bas

    #mise à jour des 2 colonnes de marge
    for i in range(1, TAILLE + 1):
        grille[i][0].en_vie = grille[i][TAILLE].en_vie #marge de gauche
        grille[i][TAILLE + 1].en_vie = grille[i][1].en_vie #marge de droite


        
def coloriage(surf, coor, envie):
    """
    coor --> tuple
    envie --> bool
    fonction qui colorie une cellule en fonction de son statut
    (noir si elle est en vie / blanche si elle est morte)
    """
    lon = 500 / TAILLE
    if envie == True :
        pygame.draw.rect(surf, (0, 0, 0), pygame.Rect((coor[0] - 1) * lon, (coor[1] - 1) * lon, lon, lon))
    elif envie == False :
        pygame.draw.rect(surf, (255, 255, 255), pygame.Rect((coor[0] - 1) * lon, (coor[1] - 1) * lon, lon, lon))
    return



def initialisation_case(surf, grille):
    """
    fonction permettant de selectionner les cases en vie en cliquant dessus
    si aucune cases n'est choisie, elle met en vie aléatoirement un certain nombre de cases
    """
    not_ready = True
    while not_ready:
        for event in pygame.event.get():
            if event.type == pygame.MOUSEBUTTONDOWN:
                if pygame.mouse.get_pressed() == (1,0,0) :
                    pos = pygame.mouse.get_pos()
                    x = pos[0]
                    y = pos[1]
                    if y < 500: #On vérifie que le click soit bien dans le tableau
                        longueure_case = 500 / TAILLE
                        x = x // longueure_case
                        y = y // longueure_case
                        # On inverse l'état de vie, on peut donc retourner à une case morte si erreur de la part de l'utilisateur
                        grille[int(x+1)][int(y+1)].en_vie = not grille[int(x+1)][int(y+1)].en_vie
                        coloriage(surf, [int(x+1), int(y+1)], grille[int(x+1)][int(y+1)].en_vie)
                        affichage_grille(surf)  #On doit réafficher car sinon on ne voit pas la grille proprement
                        pygame.display.flip()
                    if  y > 500:  #si l'utilisateur clique dans la zone en bas de la grille avec le texte
                        not_ready = False  #on sort de la boucle : ce qui va entrainer le lancement de la partie
    
    #on vérifie si l'utilisateur a selectionné une case minimum
    nev = 0
    for ligne in range(1, TAILLE + 1):
            for colonne in range(1, TAILLE + 1):
                if grille[ligne][colonne].en_vie == True:
                    nev += 1
    
    #sinon on en choisit alétoirement
    if nev == 0:
        for ligne in range(1, TAILLE + 1):
            for colonne in range(1, TAILLE + 1):
                grille[ligne][colonne].en_vie = random.randint(1,3) == 1 #une chance sur 3 d'etre vivante
                coloriage(surf, grille[ligne][colonne].position, grille[ligne][colonne].en_vie)
                affichage_grille(surf)
                pygame.display.flip()


                
                
                
                
                
                
                
                
def main():
    """
    fonction principale
    """


    nb_etapes = input("Nombre d'étapes ? ")
    #boucle pour qui ne s'arrète pas tant que l'input n'est pas un entier strictement positif
    essai = True
    while essai :
        try :
            nb_etapes = int(nb_etapes)
        except :
            nb_etapes = input("Merci d'inidiquer un entier strictement positif ")
        else :
            if 0 < nb_etapes  :
                essai = False
            else : 
                nb_etapes = input("Merci d'inidiquer un entier strictement positif ")
    
    
    
    vitesse = input("Entrer la vitesse d'éxécution s'il vous plait : \n0 pour Très rapide; 1 pour Rapide; 2 pour Normale et 3 pour Lente \n")
    #boucle pour qui ne s'arrète pas tant que l'input n'est pas un entier entre 0 et 3
    essai = True
    while essai :
        try :
            vitesse = int(vitesse)
        except :
            vitesse = input("Merci d'entrer un nombre entier entre 0 et 3 ")
        else :
            if 0<= vitesse <= 3 :
                essai = False
            else : 
                vitesse = input("Merci d'entrer un nombre entier entre 0 et 3 ")
    
    #délai du time.sleep en fonction de la vitesse choisie
    vi_dico = { 0 : 0.01, 1 : 0.5, 2 : 1, 3 : 3}
    delai = vi_dico[vitesse]

    
    run = True
    while run :
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                run = False
                pygame.quit()

        #création et affichage de la grille
        grille = creation_grille()
        affichage_grille(surf)
        
        
        #affichage du texte en dessous de la grille
        pygame.init()
        font = pygame.font.Font(None, 20)
        texte1 = font.render("Merci de sélectionner les cases en cliquant dessus", True, (255, 0, 0), (255, 255, 255))
        texte1Rect = texte1.get_rect()
        texte1Rect.center = (250,520)
        surf.blit(texte1, texte1Rect)
        
        texte2 = font.render("(Sinon l'initialisation sera aléatoire)", True, (255, 0, 0), (255, 255, 255))
        texte2Rect = texte2.get_rect()
        texte2Rect.center = (250,535)
        surf.blit(texte2, texte2Rect)
        
        texte3 = font.render("Cliquez sur ce texte pour lancer la partie", True, (255, 0, 0), (255, 255, 255))
        texte3Rect = texte3.get_rect()
        texte3Rect.center = (250,550)
        surf.blit(texte3, texte3Rect)
        
        pygame.display.flip()



        #initialisation des cellules vivantes 
        initialisation_case(surf, grille)                         

        pygame.draw.rect(surf, (255, 255, 255), pygame.Rect(0, 500, 450, 60))
        pygame.display.flip()

        for loop in range(1, nb_etapes+1):
            etape(grille)
            
            #affichage du numéro de l'étape
            font = pygame.font.Font(None, 40)
            t_etape = font.render("Etape n°" + str(loop), True, (0, 0, 0), (255, 255, 255))
            t_etapeRect = t_etape.get_rect()
            t_etapeRect.center = (250,535)
            surf.blit(t_etape, t_etapeRect)
            
            pygame.display.flip() #on met à jour l'affichage à la fin de chaque étape
            time.sleep(delai)
        
        time.sleep(5)
        run = False



    pygame.quit()
    return

main()
```
